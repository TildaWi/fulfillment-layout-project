### 2) 피킹 주문 빈도 분석

```python
# 상품별 주문 빈도 계산
product_frequency = merged_df.groupby('Product').agg({
    'Quantity': ['count', 'sum', 'mean'],
    'Location': 'nunique'
}).round(2)

product_frequency.columns = ['주문_횟수', '총_주문량', '평균_주문량', '저장_위치_수']
product_frequency = product_frequency.sort_values('주문_횟수', ascending=False).reset_index()

print("상위 10개 고빈도 상품:")
print(product_frequency.head(10))
print()
```

### 6) ABC 분류를 위한 파레토 분석

```python
# 누적 기여도 계산
product_frequency_sorted = product_frequency.sort_values('주문_횟수', ascending=False).copy()
product_frequency_sorted['누적_주문횟수'] = product_frequency_sorted['주문_횟수'].cumsum()
product_frequency_sorted['누적_비율'] = (product_frequency_sorted['누적_주문횟수'] /
                                        product_frequency_sorted['주문_횟수'].sum() * 100).round(2)
```

```python
# ABC 등급 기준점 확인
total_orders = product_frequency_sorted['주문_횟수'].sum()
a_threshold = total_orders * 0.8  # 상위 80%
b_threshold = total_orders * 0.95  # 상위 95%

a_products = len(product_frequency_sorted[product_frequency_sorted['누적_주문횟수'] <= a_threshold])
b_products = len(product_frequency_sorted[(product_frequency_sorted['누적_주문횟수'] > a_threshold) &
                                         (product_frequency_sorted['누적_주문횟수'] <= b_threshold)])
c_products = len(product_frequency_sorted) - a_products - b_products

print(f"전체 상품 수: {len(product_frequency_sorted):,}")
print(f"A등급 예상 상품 수: {a_products:,} ({a_products/len(product_frequency_sorted)*100:.1f}%)")
print(f"B등급 예상 상품 수: {b_products:,} ({b_products/len(product_frequency_sorted)*100:.1f}%)")
print(f"C등급 예상 상품 수: {c_products:,} ({c_products/len(product_frequency_sorted)*100:.1f}%)")
print()
```

```python
plt.subplot(2, 3, 6)
plt.plot(range(1, len(product_frequency_sorted)+1), product_frequency_sorted['누적_비율'])
plt.axhline(y=80, color='r', linestyle='--', label='80% 라인')
plt.axhline(y=95, color='orange', linestyle='--', label='95% 라인')
plt.title('파레토 차트 (누적 주문 비율)')
plt.xlabel('상품 순위')
plt.ylabel('누적 주문 비율 (%)')
plt.legend()

plt.tight_layout()
plt.show()
```

### 6) ABC 등급 분류

```python
# 누적 기여도 기반 ABC 분류
product_frequency_sorted = product_frequency.sort_values('주문_횟수', ascending=False).copy()
total_orders = product_frequency_sorted['주문_횟수'].sum()
product_frequency_sorted['누적_주문횟수'] = product_frequency_sorted['주문_횟수'].cumsum()
product_frequency_sorted['누적_비율'] = (product_frequency_sorted['누적_주문횟수'] / total_orders * 100)
```

```python
# ABC 분류 기준점 설정
def assign_abc_grade(row):
    if row['누적_비율'] <= 70:  # 상위 70% 기여도
        return 'A'
    elif row['누적_비율'] <= 90:  # 70-90% 기여도
        return 'B'
    else:  # 90-100% 기여도
        return 'C'

product_frequency_sorted['ABC_등급'] = product_frequency_sorted.apply(assign_abc_grade, axis=1)
```

```python
# ABC 등급별 통계
abc_stats = product_frequency_sorted.groupby('ABC_등급').agg({
    'Product': 'count',
    '주문_횟수': ['sum', 'mean', 'std'],
    '저장_위치_수': 'mean'
}).round(2)

abc_stats.columns = ['상품_수', '총_주문횟수', '평균_주문횟수', '주문횟수_표준편차', '평균_저장위치수']
abc_stats['상품_비율'] = (abc_stats['상품_수'] / len(product_frequency_sorted) * 100).round(1)
abc_stats['주문_기여도'] = (abc_stats['총_주문횟수'] / total_orders * 100).round(1)

print("ABC 등급별 상세 통계:")
print(abc_stats)
print()
```

### 3) 정규성 검정

```python
def test_normality(data, group_name):
    """정규성 검정 함수"""
    if len(data) > 5000:  # 표본이 너무 클 경우 샘플링
        data = np.random.choice(data, 5000, replace=False)

    stat, p_value = stats.shapiro(data)
    is_normal = p_value > 0.05

    print(f"{group_name}:")
    print(f"  - 통계량: {stat:.4f}, p-value: {p_value:.4f}")
    print(f"  - 정규분포 여부: {'예' if is_normal else '아니오'} (α=0.05)")
    return is_normal

# 각 등급별 정규성 검정
normality_results = {}
for grade in ['A', 'B', 'C']:
    grade_data = product_frequency_sorted[product_frequency_sorted['ABC_등급'] == grade]
    print(f"\n{grade}등급 정규성 검정:")

    # 주문 횟수 정규성 검정
    normality_results[f'{grade}_주문횟수'] = test_normality(
        grade_data['주문_횟수'].values, f"{grade}등급 주문횟수"
    )

    # 저장 위치 수 정규성 검정
    normality_results[f'{grade}_저장위치'] = test_normality(
        grade_data['저장_위치_수'].values, f"{grade}등급 저장위치수"
    )

print()
```

### 4) 분산의 동질성 검정

```python
def test_homogeneity(groups, group_names, variable_name):
    """분산 동질성 검정 함수"""
    stat, p_value = stats.levene(*groups)
    is_homogeneous = p_value > 0.05

    print(f"{variable_name} 분산 동질성 검정:")
    print(f"  - Levene 통계량: {stat:.4f}, p-value: {p_value:.4f}")
    print(f"  - 분산 동질성: {'만족' if is_homogeneous else '불만족'} (α=0.05)")
    return is_homogeneous

# 등급별 그룹 데이터 추출
a_grade = product_frequency_sorted[product_frequency_sorted['ABC_등급'] == 'A']
b_grade = product_frequency_sorted[product_frequency_sorted['ABC_등급'] == 'B']
c_grade = product_frequency_sorted[product_frequency_sorted['ABC_등급'] == 'C']

# 주문 횟수 분산 동질성 검정
order_homogeneity = test_homogeneity(
    [a_grade['주문_횟수'], b_grade['주문_횟수'], c_grade['주문_횟수']],
    ['A등급', 'B등급', 'C등급'],
    '주문 횟수'
)

# 저장 위치 수 분산 동질성 검정
location_homogeneity = test_homogeneity(
    [a_grade['저장_위치_수'], b_grade['저장_위치_수'], c_grade['저장_위치_수']],
    ['A등급', 'B등급', 'C등급'],
    '저장 위치 수'
)

print()
```

### 5) 일원분산분석

```python
def perform_anova(groups, group_names, variable_name):
    """일원분산분석 수행 함수"""
    f_stat, p_value = f_oneway(*groups)

    print(f"{variable_name} ANOVA 결과:")
    print(f"  - F-통계량: {f_stat:.4f}")
    print(f"  - p-value: {p_value:.4f}")

    if p_value < 0.001:
        significance = "p < 0.001 (매우 유의함)"
    elif p_value < 0.01:
        significance = "p < 0.01 (유의함)"
    elif p_value < 0.05:
        significance = "p < 0.05 (유의함)"
    else:
        significance = "p ≥ 0.05 (유의하지 않음)"

    print(f"  - 유의성: {significance}")

    if p_value < 0.05:
        print(f"  - 결론: ABC 등급 간 {variable_name}에 유의한 차이가 있음")
    else:
        print(f"  - 결론: ABC 등급 간 {variable_name}에 유의한 차이가 없음")

    return f_stat, p_value

# 주문 횟수 ANOVA
print("주문 횟수에 대한 등급 간 차이 검정:")
order_f_stat, order_p_value = perform_anova(
    [a_grade['주문_횟수'], b_grade['주문_횟수'], c_grade['주문_횟수']],
    ['A등급', 'B등급', 'C등급'],
    '주문 횟수'
)

print()

# 저장 위치 수 ANOVA
print("저장 위치 수에 대한 등급 간 차이 검정:")
location_f_stat, location_p_value = perform_anova(
    [a_grade['저장_위치_수'], b_grade['저장_위치_수'], c_grade['저장_위치_수']],
    ['A등급', 'B등급', 'C등급'],
    '저장 위치 수'
)

print()
```

### 6) 사후 검정

```python
from scipy.stats import tukey_hsd

def perform_tukey_hsd(groups, group_names, variable_name):
    """Tukey HSD 사후 검정 수행"""
    print(f"{variable_name} Tukey HSD 사후 검정:")

    # Tukey HSD 검정 수행
    tukey_result = tukey_hsd(*groups)

    # 결과 해석
    comparisons = [(0, 1, 'A vs B'), (0, 2, 'A vs C'), (1, 2, 'B vs C')]

    for i, j, comparison in comparisons:
        p_val = tukey_result.pvalue[i, j]
        mean_diff = np.mean(groups[i]) - np.mean(groups[j])

        if p_val < 0.001:
            significance = "p < 0.001 (매우 유의함)"
        elif p_val < 0.01:
            significance = "p < 0.01 (유의함)"
        elif p_val < 0.05:
            significance = "p < 0.05 (유의함)"
        else:
            significance = "p ≥ 0.05 (유의하지 않음)"

        print(f"  - {comparison}: 평균차이 = {mean_diff:.2f}, {significance}")

# 주문 횟수 사후 검정
perform_tukey_hsd(
    [a_grade['주문_횟수'], b_grade['주문_횟수'], c_grade['주문_횟수']],
    ['A등급', 'B등급', 'C등급'],
    '주문 횟수'
)

print()

# 저장 위치 수 사후 검정
perform_tukey_hsd(
    [a_grade['저장_위치_수'], b_grade['저장_위치_수'], c_grade['저장_위치_수']],
    ['A등급', 'B등급', 'C등급'],
    '저장 위치 수'
)

print()
```

### 7) 효과 크기 검정

```python
def calculate_eta_squared(f_stat, df_between, df_within):
    """에타 제곱 효과 크기 계산"""
    eta_squared = (f_stat * df_between) / (f_stat * df_between + df_within)
    return eta_squared

def interpret_eta_squared(eta_squared):
    """에타 제곱 해석"""
    if eta_squared < 0.01:
        return "매우 작음"
    elif eta_squared < 0.06:
        return "작음"
    elif eta_squared < 0.14:
        return "중간"
    else:
        return "큼"

# 주문 횟수 효과 크기
n_total = len(product_frequency_sorted)
n_groups = 3
df_between = n_groups - 1
df_within = n_total - n_groups

order_eta_squared = calculate_eta_squared(order_f_stat, df_between, df_within)
location_eta_squared = calculate_eta_squared(location_f_stat, df_between, df_within)

print(f"주문 횟수 효과 크기:")
print(f"  - Eta² = {order_eta_squared:.4f}")
print(f"  - 해석: {interpret_eta_squared(order_eta_squared)}")
print()

print(f"저장 위치 수 효과 크기:")
print(f"  - Eta² = {location_eta_squared:.4f}")
print(f"  - 해석: {interpret_eta_squared(location_eta_squared)}")
print()
```

### 8) 시뮬레이션

```python
# ABC 최적화 시나리오 설정
def simulate_abc_optimization(df):
    """ABC 최적화 시나리오 시뮬레이션"""
    df_optimized = df.copy()

    # ABC 등급별 최적화 정책
    optimization_policy = {
        'A': 1,    # A등급: 1개 위치 집중
        'B': 2,    # B등급: 2개 위치
        'C': None  # C등급: 현재 유지
    }

    for grade in ['A', 'B']:
        mask = df_optimized['ABC_등급'] == grade
        df_optimized.loc[mask, 'ABC_최적화_위치수'] = optimization_policy[grade]

    # C등급은 현재 위치 수 유지
    c_mask = df_optimized['ABC_등급'] == 'C'
    df_optimized.loc[c_mask, 'ABC_최적화_위치수'] = df_optimized.loc[c_mask, '저장_위치_수']

    return df_optimized

# 시뮬레이션 실행
product_optimized = simulate_abc_optimization(product_frequency_sorted)

# 개선 효과 계산
print("ABC 최적화 효과 분석:")

for grade in ['A', 'B', 'C']:
    grade_data = product_optimized[product_optimized['ABC_등급'] == grade]

    current_avg = grade_data['저장_위치_수'].mean()
    optimized_avg = grade_data['ABC_최적화_위치수'].mean()
    improvement = (current_avg - optimized_avg) / current_avg * 100 if current_avg > optimized_avg else 0

    print(f"\n{grade}등급 ({len(grade_data)}개 상품):")
    print(f"  - 현재 평균 저장 위치 수: {current_avg:.2f}개")
    print(f"  - 최적화 후 저장 위치 수: {optimized_avg:.2f}개")
    print(f"  - 개선율: {improvement:.1f}%")

# 전체 가중 평균 개선 효과
total_orders = product_optimized['주문_횟수'].sum()
weighted_improvement = 0

for grade in ['A', 'B', 'C']:
    grade_data = product_optimized[product_optimized['ABC_등급'] == grade]
    grade_contribution = grade_data['주문_횟수'].sum() / total_orders

    current_avg = grade_data['저장_위치_수'].mean()
    optimized_avg = grade_data['ABC_최적화_위치수'].mean()
    grade_improvement = (current_avg - optimized_avg) / current_avg * 100 if current_avg > optimized_avg else 0

    weighted_improvement += grade_contribution * grade_improvement

print(f"\n 전체 가중평균 개선 효과: {weighted_improvement:.1f}%")
```

### 9) t-검정

```python
# A등급과 B등급에 대해서만 개선이 있으므로 이들을 대상으로 t-검정 수행
for grade in ['A', 'B']:
    grade_data = product_optimized[product_optimized['ABC_등급'] == grade]

    current_locations = grade_data['저장_위치_수']
    optimized_locations = grade_data['ABC_최적화_위치수']

    # 대응표본 t-검정 수행
    t_stat, p_value = ttest_rel(current_locations, optimized_locations)

    print(f"\n{grade}등급 개선 효과 t-검정:")
    print(f"  - t-통계량: {t_stat:.4f}")
    print(f"  - p-value: {p_value:.4f}")

    if p_value < 0.001:
        significance = "p < 0.001 (매우 유의함)"
    elif p_value < 0.01:
        significance = "p < 0.01 (유의함)"
    elif p_value < 0.05:
        significance = "p < 0.05 (유의함)"
    else:
        significance = "p ≥ 0.05 (유의하지 않음)"

    print(f"  - 유의성: {significance}")
    print(f"  - 결론: ABC 최적화가 {grade}등급 상품의 위치 집중도를 {'유의하게 개선함' if p_value < 0.05 else '유의하게 개선하지 못함'}")
```

## **6.결과 분석**

### 1) 센터 구역 정의

```python
# 센터 구역 정의
warehouse_zones = {
    'A': {'priority': 1, 'distance_from_entrance': 10, 'accessibility': 'high'},
    'B': {'priority': 2, 'distance_from_entrance': 25, 'accessibility': 'medium'},
    'C': {'priority': 3, 'distance_from_entrance': 40, 'accessibility': 'low'}
}

np.random.seed(42)

df_sorted['current_zone'] = np.random.choice(['A', 'B', 'C'], size=len(df_sorted))
df_sorted['optimal_zone'] = df_sorted['ABC_등급']
```

### 2) 피킹 효율성 계산

```python
def calculate_picking_metrics(df, zone_column):
    """피킹 효율성 지표 계산"""

    # 주문_횟수 컬럼 확인 및 대체
    freq_column = '주문_횟수' if '주문_횟수' in df.columns else 'order_frequency'

    # 평균 이동 거리 계산
    df['zone_distance'] = df[zone_column].map(
        lambda x: warehouse_zones[x]['distance_from_entrance']
    )
    avg_travel_distance = (df['zone_distance'] * df[freq_column]).sum() / df[freq_column].sum()

    # 피킹 작업 시간 (거리 기반)
    base_picking_time = 2  # 기본 피킹 시간 (분)
    travel_time_per_meter = 0.1  # 미터당 이동 시간 (분)

    df['picking_time'] = base_picking_time + (df['zone_distance'] * travel_time_per_meter)
    avg_picking_time = (df['picking_time'] * df[freq_column]).sum() / df[freq_column].sum()

    # 시간당 피킹 수량
    picks_per_hour = 60 / avg_picking_time

    return {
        'avg_travel_distance': avg_travel_distance,
        'avg_picking_time': avg_picking_time,
        'picks_per_hour': picks_per_hour
    }

# 기존 프로세스 vs ABC 등급 기반 프로세스 비교
current_metrics = calculate_picking_metrics(df_sorted, 'current_zone')
optimal_metrics = calculate_picking_metrics(df_sorted, 'optimal_zone')

print(" 피킹 효율성 지표:")
print(f"\n1. 피킹 효율성 지표:")
print(f"   • 평균 이동 거리 (Average Travel Distance)")
print(f"     - 기존 프로세스: {current_metrics['avg_travel_distance']:.1f}m")
print(f"     - ABC 등급 기반: {optimal_metrics['avg_travel_distance']:.1f}m")
print(f"     - 개선율: {((current_metrics['avg_travel_distance'] - optimal_metrics['avg_travel_distance']) / current_metrics['avg_travel_distance'] * 100):.1f}%")

print(f"\n   • 피킹 작업 시간 (Picking Time)")
print(f"     - 기존 프로세스: {current_metrics['avg_picking_time']:.2f}분")
print(f"     - ABC 등급 기반: {optimal_metrics['avg_picking_time']:.2f}분")
print(f"     - 개선율: {((current_metrics['avg_picking_time'] - optimal_metrics['avg_picking_time']) / current_metrics['avg_picking_time'] * 100):.1f}%")

print(f"\n   • 시간당 피킹 수량 (Picks per Hour)")
print(f"     - 기존 프로세스: {current_metrics['picks_per_hour']:.1f}개")
print(f"     - ABC 등급 기반: {optimal_metrics['picks_per_hour']:.1f}개")
print(f"     - 개선율: {((optimal_metrics['picks_per_hour'] - current_metrics['picks_per_hour']) / current_metrics['picks_per_hour'] * 100):.1f}%")
```

### 3) 진열 효율성 계산

```python
def calculate_layout_metrics(df, zone_column):
    """진열 효율성 지표 계산"""

    # 주문_횟수 컬럼 확인 및 대체
    freq_column = '주문_횟수' if '주문_횟수' in df.columns else 'order_frequency'

    # 고빈도 상품 접근성 점수
    high_freq_threshold = df[freq_column].quantile(0.8)
    high_freq_products = df[df[freq_column] >= high_freq_threshold]

    high_freq_in_zone_a = len(high_freq_products[high_freq_products[zone_column] == 'A'])
    accessibility_score = (high_freq_in_zone_a / len(high_freq_products)) * 100

    # 위치 집중도 지수 (같은 등급 상품들이 얼마나 집중되어 있는지)
    grade_counts = df.groupby([df['ABC_등급'], zone_column]).size().unstack(fill_value=0)
    total_by_grade = grade_counts.sum(axis=1)

    concentration_scores = []
    for grade in ['A', 'B', 'C']:
        if grade in grade_counts.index:
            max_in_optimal_zone = grade_counts.loc[grade, grade] if grade in grade_counts.columns else 0
            concentration = (max_in_optimal_zone / total_by_grade[grade]) * 100
            concentration_scores.append(concentration)

    location_concentration_index = np.mean(concentration_scores)

    # 공간 활용률 (각 구역의 활용도)
    zone_utilization = df.groupby(zone_column).size() / len(df) * 100
    space_utilization_rate = np.std(zone_utilization)  # 균등한 분산이 좋음 (낮을수록 좋음)

    return {
        'accessibility_score': accessibility_score,
        'location_concentration_index': location_concentration_index,
        'space_utilization_rate': space_utilization_rate
    }

# 진열 효율성 비교
current_layout = calculate_layout_metrics(df_sorted, 'current_zone')
optimal_layout = calculate_layout_metrics(df_sorted, 'optimal_zone')

print(" 진열 효율성 지표:")
print(f"\n2. 진열 효율성 지표:")
print(f"   • 고빈도 상품 접근성 점수 (High-frequency Product Accessibility Score)")
print(f"     - 기존 프로세스: {current_layout['accessibility_score']:.1f}%")
print(f"     - ABC 등급 기반: {optimal_layout['accessibility_score']:.1f}%")
print(f"     - 개선: +{(optimal_layout['accessibility_score'] - current_layout['accessibility_score']):.1f}%p")

print(f"\n   • 위치 집중도 지수 (Location Concentration Index)")
print(f"     - 기존 프로세스: {current_layout['location_concentration_index']:.1f}%")
print(f"     - ABC 등급 기반: {optimal_layout['location_concentration_index']:.1f}%")
print(f"     - 개선: +{(optimal_layout['location_concentration_index'] - current_layout['location_concentration_index']):.1f}%p")

print(f"\n   • 공간 활용률 (Space Utilization Rate)")
print(f"     - 기존 프로세스: {current_layout['space_utilization_rate']:.1f}")
print(f"     - ABC 등급 기반: {optimal_layout['space_utilization_rate']:.1f}")
print(f"     - 개선율: {((current_layout['space_utilization_rate'] - optimal_layout['space_utilization_rate']) / current_layout['space_utilization_rate'] * 100):.1f}%")
```

### 4) 운영 효율성 계산

```python
def calculate_operational_metrics(df, zone_column, picking_metrics):
    """운영 효율성 지표 계산"""

    # 주문_횟수 컬럼 확인 및 대체
    freq_column = '주문_횟수' if '주문_횟수' in df.columns else 'order_frequency'

    # 전체 처리 시간 (Total Processing Time)
    total_orders = df[freq_column].sum()
    total_processing_time = total_orders * picking_metrics['avg_picking_time']

    # 작업자 생산성 (Worker Productivity)
    worker_productivity = picking_metrics['picks_per_hour']

    # 오류율 (Error Rate) - 거리와 복잡성에 비례
    base_error_rate = 0.02  # 2% 기본 오류율
    distance_factor = picking_metrics['avg_travel_distance'] / 40  # 최대 거리 40m 기준
    error_rate = base_error_rate * (1 + distance_factor)

    return {
        'total_processing_time': total_processing_time,
        'worker_productivity': worker_productivity,
        'error_rate': error_rate * 100
    }

# 운영 효율성 비교
current_operational = calculate_operational_metrics(df_sorted, 'current_zone', current_metrics)
optimal_operational = calculate_operational_metrics(df_sorted, 'optimal_zone', optimal_metrics)

print(" 운영 효율성 지표:")
print(f"\n3. 운영 효율성 지표:")
print(f"   • 전체 처리 시간 (Total Processing Time)")
print(f"     - 기존 프로세스: {current_operational['total_processing_time']:.0f}분")
print(f"     - ABC 등급 기반: {optimal_operational['total_processing_time']:.0f}분")
print(f"     - 개선율: {((current_operational['total_processing_time'] - optimal_operational['total_processing_time']) / current_operational['total_processing_time'] * 100):.1f}%")

print(f"\n   • 작업자 생산성 (Worker Productivity)")
print(f"     - 기존 프로세스: {current_operational['worker_productivity']:.1f}개/시간")
print(f"     - ABC 등급 기반: {optimal_operational['worker_productivity']:.1f}개/시간")
print(f"     - 개선율: {((optimal_operational['worker_productivity'] - current_operational['worker_productivity']) / current_operational['worker_productivity'] * 100):.1f}%")

print(f"\n   • 오류율 (Error Rate)")
print(f"     - 기존 프로세스: {current_operational['error_rate']:.2f}%")
print(f"     - ABC 등급 기반: {optimal_operational['error_rate']:.2f}%")
print(f"     - 개선: -{(current_operational['error_rate'] - optimal_operational['error_rate']):.2f}%p")
```

### 5) 지표 개선 요약
